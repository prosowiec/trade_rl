```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO
```

```python
prices_df = pd.read_csv('test.csv')
prices_df['Date'] = pd.to_datetime(prices_df['Date'])

trades_df = pd.read_csv('trades.csv')
trades_df['Date'] = pd.to_datetime(trades_df['timestamp'])
trades_df

trades_df
```

```python
def plot_trader_vs_prices(prices_df, trades_df, selected_asset=None):
    """
    Wizualizacja działań tradera (BUY / SELL / HOLD) na tle wykresu ceny aktywa.
    Nie interpoluje braków danych (NaN).
    """
    from matplotlib.ticker import FuncFormatter
    import numpy as np
    from matplotlib.lines import Line2D
    
    def equidate_ax(fig, ax, dates, fmt="%m-%d %H:%M", label="Czas"):
        """
        Sets all relevant parameters for an equidistant date-x-axis.
        """    
        N = len(dates)
        def format_date(index, pos):
            index = np.clip(int(index + 0.5), 0, N - 1)
            return dates[index].strftime(fmt)
        ax.xaxis.set_major_formatter(FuncFormatter(format_date))
        ax.set_xlabel(label)
        fig.autofmt_xdate()
    
    print("Trades DataFrame:")
    display(trades_df)
    
    if prices_df.empty or trades_df.empty:
        print("⚠️ Brak danych do wizualizacji.")
        return
    
    # --- Przygotowanie danych cenowych ---
    if "Date" in prices_df.columns:
        prices_df["Date"] = pd.to_datetime(prices_df["Date"])
        prices_df = prices_df.set_index("Date")
    prices_df = prices_df.sort_index()
    
    # --- Przygotowanie danych transakcji ---
    trades_df = trades_df.copy()
    trades_df["timestamp"] = pd.to_datetime(trades_df["timestamp"])
    
    # --- Lista aktywów ---
    assets = [selected_asset] if selected_asset else prices_df.columns
    
    for asset in assets:
        if asset not in prices_df.columns:
            print(f"⚠️ Brak danych dla {asset}")
            continue
        
        # Dane cenowe dla danego aktywa (bez NaN)
        asset_prices = prices_df[asset].dropna()
        
        if asset_prices.empty:
            print(f"⚠️ Brak danych cenowych dla {asset}")
            continue
        
        # --- Wykres ---
        fig, ax = plt.subplots(figsize=(12, 5))
        
        # Przygotowanie indeksów dla equidistant plot
        dates_list = asset_prices.index.tolist()
        x_indices = np.arange(len(dates_list))
        
        # CENA - wykres z equidistant spacing
        ax.plot(
            x_indices,
            asset_prices.values,
            label="Cena",
            color="black",
            linewidth=1.5,
        )
        
        # Transakcje dla tego aktywa
        asset_trades = trades_df[trades_df["asset_name"] == asset].copy()
        
        print(f"\n{asset} - Found {len(asset_trades)} trades")
        
        if not asset_trades.empty:
            # BUY
            buy_trades = asset_trades[asset_trades["action"] == "BUY"]
            print(f"  BUY trades: {len(buy_trades)}")
            
            if not buy_trades.empty:
                for idx, trade in buy_trades.iterrows():
                    try:
                        trade_time = trade["timestamp"]
                        # Znajdź najbliższy indeks w asset_prices
                        # Używamy get_indexer z method='nearest'
                        loc = asset_prices.index.get_indexer([trade_time], method='nearest')[0]
                        
                        if loc >= 0 and loc < len(dates_list):
                            price = asset_prices.iloc[loc]
                            print(f"    BUY at {trade_time} -> index {loc}, price {price}")
                            ax.scatter(
                                loc,
                                price,
                                color="green", marker="^", s=150, zorder=5, edgecolors='darkgreen', linewidths=1.5
                            )
                    except Exception as e:
                        print(f"    Error plotting BUY: {e}")
            
            # SELL
            sell_trades = asset_trades[asset_trades["action"] == "SELL"]
            print(f"  SELL trades: {len(sell_trades)}")
            
            if not sell_trades.empty:
                for idx, trade in sell_trades.iterrows():
                    try:
                        trade_time = trade["timestamp"]
                        # Znajdź najbliższy indeks w asset_prices
                        loc = asset_prices.index.get_indexer([trade_time], method='nearest')[0]
                        
                        if loc >= 0 and loc < len(dates_list):
                            price = asset_prices.iloc[loc]
                            print(f"    SELL at {trade_time} -> index {loc}, price {price}")
                            ax.scatter(
                                loc,
                                price,
                                color="red", marker="v", s=150, zorder=5, edgecolors='darkred', linewidths=1.5
                            )
                    except Exception as e:
                        print(f"    Error plotting SELL: {e}")
        
        # Legenda
        legend_elements = [
            Line2D([0], [0], color='black', linewidth=1.5, label='Cena')
        ]
        if len(asset_trades[asset_trades["action"] == "BUY"]) > 0:
            legend_elements.append(Line2D([0], [0], marker='^', color='w', 
                                         markerfacecolor='g', markersize=10, label='BUY'))
        if len(asset_trades[asset_trades["action"] == "SELL"]) > 0:
            legend_elements.append(Line2D([0], [0], marker='v', color='w', 
                                         markerfacecolor='r', markersize=10, label='SELL'))
        
        ax.legend(handles=legend_elements)
        ax.set_title(f"Działania tradera dla {asset}")
        ax.set_ylabel("Cena")
        ax.grid(True)
        
        # Zastosuj equidistant date formatting
        equidate_ax(fig, ax, dates_list)
        
        plt.tight_layout()
        plt.show()
        
plot_trader_vs_prices(prices_df, trades_df)
```
